# Matplotlib

## 绘制正弦曲线

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.arange(0, 2 * np.pi, 0.01)    # 以 0.01 为单位，生成 0 到 2pi 的数据
y = np.sin(x)

# 绘制图形
plt.plot(x, y)
plt.show()
```

## 绘制三维散点图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个 3D 图形对象
figure = plt.figure()
axes = plt.axes(projection='3d')

# 生成随机数据
x = np.random.normal(size=200)
y = np.random.normal(size=200)
z = np.random.normal(size=200)

# 绘制散点图
axes.scatter(x, y, z)

# 设置坐标轴标签
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_zlabel('Z')

# 显示图形
plt.show()
```

## 绘制二元高斯函数

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个 3D 图形对象
figure = plt.figure(figsize=(12, 6))
axes = plt.axes(projection="3d")

# 生成数据
x, y = np.mgrid[-3:3:100j, -3:3:100j]       # 在 x 和 y 方向上都创建一个范围为 -3 到 3 的网格，其中各包含 100 个点
z = np.exp(-x ** 2 - y ** 2)                # z 为二元高斯函数的值
                                            # f(x, y) = A * exp(-(x - x0) ** 2 / (2 * sigma_x ** 2) - (y - y0) ** 2 / (2 * sigma_y ** 2))
                                            # A 是函数的幅度，x0 和 y0 是函数的中心位置，sigma_x 和 sigma_y 是函数的标准差
                                            # 二元高斯函数通常用于图像处理和计算机视觉中的滤波和特征提取

# 创建三维表面图
surface = axes.plot_surface(x, y, z,                            # x、y 和 z 分别是三个数组，表示表面图的三个坐标轴
                            rstride=1,                          # rstride 和 cstride 分别表示行和列的步长，用于控制表面图的密度
                            cstride=1,
                            cmap=plt.get_cmap('rainbow'))       # cmap 表示颜色映射，用于控制表面图的颜色

# 设置坐标轴
axes.set_zlim(-0.1, 1)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_zlabel("Z")

# 显示图形
figure.colorbar(surface, shrink=0.5, aspect=5)      # 创建一个颜色条，surface 是一个 plot_surface 函数的返回值，表示要添加颜色条的三维表面图
                                                    # shrink 和 aspect 分别表示颜色条的缩放比例和宽高比
plt.show()
```

## 绘制三维曲线图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个 3D 图形对象
figure = plt.figure(figsize=(12, 6))
axes = plt.axes(projection="3d")

# 生成数据
theta = np.linspace(0, np.pi * 10, 1000)    # 生成了一个包含 1000 个元素的等差数列 theta (极角)
r = np.linspace(0, 1, 1000)                 # 生成了一个包含 1000 个元素的等差数列 r (极径)
x = r * np.cos(theta)                       # 计算极坐标系中的点 x y
y = r * np.sin(theta)
z = np.linspace(0, 4, 1000)

# 绘制三维线图
axes.plot(x, y, z, color='gold', lw=3)      # color 参数控制线图的颜色，lw 参数控制线图的线宽

# 显示图形
plt.show()
```
